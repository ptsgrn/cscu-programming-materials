**Recursion**

````python
def fac (n) :
    if n == 0 : # basis case
        return 1
    else :
        return n * fac(n-1) # call function
````

````python
fac(5)
````

````python
def fib (n) :
    if n == 1 or n == 0 : # basis case
        return 1
    else :
        return fib(n-1) + fib(n-2) # call function
````

````python
print(fib(0))
print(fib(1))
print(fib(2))
print(fib(3))
print(fib(4))
print(fib(5))
print(fib(6))
print(fib(7))
````
