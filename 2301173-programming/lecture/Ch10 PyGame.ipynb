## The Window Coordinate System
image.png

*   the origin pixel is the point(0,0) --> the upper-left corner of the window.
*   x values (width) and y values (height) start at zero at the left and top of the window, respectively
*   The x and y values are always integers. Each (x, y) pair specifies a single
pixel in the window.



### Pixel Colors
*   Each pixel on the screen is made up of a combination of three colors: red,
green, and blue (RGB).
*   A value from 0, meaning none, to 255, meaning full intensity. Therefore, there are 256 x 256 x 256 possible combinations, or 16,777,216 (16 million) possible colors for each pixel.

Colors in pygame writed as Python tuples of three numbers.

## Event

Something that happens while your program is running that
your program wants to or needs to respond to. Most events
are generated by user actions.

An event-driven GUI program runs constantly in an infinite loop. Each time through the loop, the program checks for any new events
*   needs to react and execute appropriate code to handle those events.
*   needs to redraw all the elements in the window to update what the user sees.
https://www.pygame.org/docs/ref/event.html

````python
# Import the pygame library
import pygame

# Import system
import sys

# Initialize the pygame library
pygame.init()

# Set up the drawing window with (WINDOW_WIDTH,WINDOW_HEIGHT)
screen = pygame.display.set_mode((500,500))

# Set up the drawing window with "title"
pygame.display.set_caption("First PyGame")

# Define backgound color
background = pygame.Color(0, 0, 0)

# Run until the user asks to quit
running = True
while running:
    # Detect all events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_r:
                background = pygame.Color('red') # (255, 0, 0)
            elif event.key == pygame.K_g:
                background = pygame.Color('green') # (0, 255, 0)
            elif event.key == pygame.K_b:
                background = pygame.Color('blue') # (0, 0, 255)
            elif event.key == pygame.K_ESCAPE:
                running = False

    # Fill the screen background
    screen.fill(background)

    # Update the screen
    pygame.display.flip()

pygame.quit()
sys.exit()
````

````python
# Import the pygame library
import pygame

# Import system
import sys

# Initialize the pygame library
pygame.init()

# Set up the drawing window with (WINDOW_WIDTH,WINDOW_HEIGHT)
screen = pygame.display.set_mode((500,500))

# Set up the drawing window with "title"
pygame.display.set_caption("First PyGame")

# Define backgound color
background = pygame.Color(0, 0, 0)

# Load image
ballImage = pygame.image.load('ball.png')
pointerImage = pygame.image.load('pencil.jpg')

# Assign pointer and ball position
pointerX = 100
pointerY = 100

ballRect = pygame.Rect(300, 300, ballImage.get_width(), ballImage.get_height())

# Run until the user asks to quit
running = True
while running:
    # Detect all events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                pointerX = pointerX - 10
            elif event.key == pygame.K_RIGHT:
                pointerX = pointerX + 10
            elif event.key == pygame.K_UP:
                pointerY = pointerY - 10
            elif event.key == pygame.K_DOWN:
                pointerY = pointerY + 10

    pointerRect = pygame.Rect(pointerX, pointerY, pointerImage.get_width(), pointerImage.get_height())

    # Check  collision
    if pointerRect.colliderect(ballRect):
         pointerX = pointerX - 100

    # Fill the screen background
    screen.fill(background)

    # draw ball at position x = 300, y = 300
    screen.blit(ballImage, (300, 300))
    screen.blit(pointerImage, (pointerX, pointerY))

    # Update the screen
    pygame.display.update()

pygame.quit()
sys.exit()
````

### `pygame.Rect`
pygame object for storing rectangular coordinates. https://www.pygame.org/docs/ref/rect.html
image.png

````python
# Import the pygame library
import pygame

# Import system
import sys

# Initialize the pygame library
pygame.init()

# Set up the drawing window with (WINDOW_WIDTH,WINDOW_HEIGHT)
screen = pygame.display.set_mode((500,500))

# Set up the drawing window with "title"
pygame.display.set_caption("First PyGame")

# Define backgound color
background = pygame.Color(0, 0, 0)

# Load image
ballImage = pygame.image.load('ball.png')

# ballRect = pygame.Rect(ballX, ballY, ballImage.get_width(), ballImage.get_height())
ballRect = ballImage.get_rect()

ballRect.centerx = 250
ballRect.centery = 250

ballSpeed = 2

# Run until the user asks to quit
running = True
while running:
    # Detect all events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Get mouse position
    mouseX, mouseY = pygame.mouse.get_pos()

    # Caculate vector distance
    dx = mouseX - ballRect.centerx
    dy = mouseY - ballRect.centery

    distance = (dx ** 2 + dy ** 2) ** 0.5

    # Caculate time to reach the destination
    time_to_reach = distance / ballSpeed

    if time_to_reach > 0:
        vx = dx / time_to_reach
        vy = dy / time_to_reach

        # Move ball
        ballRect.x += vx
        ballRect.y+= vy

    # Fill the screen background
    screen.fill(background)

    # draw ball
    screen.blit(ballImage, ballRect)

    # Update the screen
    pygame.display.update()

pygame.quit()
sys.exit()
````

## Sprite
pygame module with basic game object classes

https://www.pygame.org/docs/ref/sprite.html

````python
# http://www.codingwithruss.com/pygame/sprite-class-and-sprite-groups-explained/

# Import the pygame library
import pygame

import random

# Initialize the pygame library
pygame.init()

# Creat clock
clock = pygame.time.Clock()

# Set up the drawing window with (WINDOW_WIDTH,WINDOW_HEIGHT)
screen = pygame.display.set_mode((500,500))

# Set up the drawing window with "title"
pygame.display.set_caption("First PyGame")

# Define backgound color
background = pygame.Color('aqua')

# Colours list
colours = ['magenta', 'silver', 'coral', 'darkorange', 'forestgreen', 'lime', 'navy']

# Create class for squares
class Square(pygame.sprite.Sprite):
  def __init__(self, colr, x, y):
    pygame.sprite.Sprite.__init__(self)
    self.image = pygame.Surface((20, 20))
    self.image.fill(colr)
    self.rect = self.image.get_rect()
    self.rect.center = (x, y)

  def update(self):
    self.rect.move_ip(0, 5)

    # Check if sprite has gone off screen
    if self.rect.top > 500:
      self.kill()

# Create sprite group for squares
squares = pygame.sprite.Group()

#create square and add to squares group
square = Square("crimson", 250, 250)
squares.add(square)

# Run until the user asks to quit
running = True
while running:

   # Detect all events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            # Get mouse coordinates
            pos = pygame.mouse.get_pos()
            # Create square
            square = Square(random.choice(colours), pos[0], pos[1])
            squares.add(square)

    clock.tick(60)

    # Fill the screen background
    screen.fill(background)

    # Update sprite group
    squares.update()

    # Draw sprite group
    squares.draw(screen)

    # Update the screen
    pygame.display.flip()


pygame.quit()

````
