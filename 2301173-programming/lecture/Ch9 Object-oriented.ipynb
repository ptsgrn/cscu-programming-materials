## Inheritance


*   Allows subclass/child class to inherit the properties of a class from superclass/parent class
*   Base class vs. Derived class
*   Represents real-world relationships well. "is-a" relationship between classes, where the subclass inherits the attributes and behaviors (properties and methods) of the superclass
*   Provides the reusability of a code.
*   If class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.
*   

```
#create class
class SubclassName(SuperclassName):
#create object
objectName = SubclassName(attribute, ..., attribute)
```

## Polymorphism
*   Different objects of different classes can be treated as objects of a common superclass.
*   To demonstrate polymorphism through method overriding or overloading, which allows sub-classes to provide a specific implementation of a method that is already defined in a superclass.
*   Polymorphism simplifies code, enhances code reusability, and allows for more generic and flexible design.
*   Overriding - occurs at runtime and allows a subclass to provide a specific implementation of a method that is already defined in its superclass --> based on the actual type of the object at runtime.
*   Overloading - occurs at compile time and involves multiple methods in the same class with the same name but different parameters --> based on the method's signature.



image.png

````python
class Shape:
    def __init__(self, shapeName):
        self.shapeName = shapeName
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        Shape.__init__(self, shapeName= 'Circle')
        self.radius = radius
    def area(self):
        return 3.14159265359 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        Shape.__init__(self, shapeName='Rectangle')
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height

circle1 = Circle(5)
rectangle1 = Rectangle(4, 6)
circle2 = Circle(3)
rectangle2 = Rectangle(2, 3)

shapes = [circle1, rectangle1, circle2, rectangle2]

for shape in shapes:
   print(shape.shapeName, 'area =', shape.area())
````

## Composition
*   Involves creating objects of one class within another class to build complex objects and define relationships between them.
*   Allows to reuse and combine existing classes as components to create more sophisticated and flexible objects, without relying on class inheritance.



image.png

````python
class Room:
  def __init__(self, id, num_seat):
    self.id = id
    self.num_seat = num_seat
  def __str__(self): # return class object
    return f"{self.id} {self.num_seat}"

class Building:
  def __init__(self, name, num_room, min_seat):
    self.name = name
    self.num_room  = num_room
    self.room = {}
    for i in range (num_room):
      self.room[i] = Room(i+1, min_seat)
      min_seat += 5

b1 = Building("MHMK", 10, 10)
print(b1.name)
for i in range(10):
  print(b1.room[i])
````

````python
b2 = Building("MHVH", 10, 10)
print(b2.name)
for i in range(10):
  print(b2.room[i])
````

## Aggregation
.  In Python, aggregation is implemented by allowing


*  Represents a "whole-part" relationship between classes, where one class (the whole) contains or is composed of other classes (the parts).
*  Unlike composition, which implies a strong ownership relationship, aggregation typically represents a weaker form of association.
*  One class to have references to instances of another class, often referred to as objects or components.

## Encapsulation
*   Refers to the bundling of data (attributes) and method (functions) that operate on the data into a single unit, known as a class
*   Enforces data hiding and restricts direct access to certain aspects of an object, making it possible to control and manage the state of an object.


Public: Attributes and methods are accessible from outside the class without any restrictions. (e.g., self.variable or def method(self)).

Protected: Attributes and methods are intended to be used within the class or its subclasses. (e.g., _variable or _method()).

Private: Attributes and methods are meant to be kept within the class and not accessed from outside the class. (e.g., __variable or __method()).

However, Python relies on conventions rather than strict access control, so it's still possible to access protected and private members if know the naming conventions or use the name mangling syntax.

````python
class Car:
    def __init__(self, brand, model):
        self._brand = brand  # Protected attribute (convention)
        self.__model = model  # Private attribute

    # Public method
    def get_brand(self):
        return self._brand
    def set_brand(self, brand):
        self._brand = brand

    # Private method
    def __get_model(self):
        return self.__model

car1 = Car("Toyota", "SUV")

# Access the protected attribute (convention)
print(car1._brand)
print(car1.get_brand())
````

````python
print(car1.__model)
````

````python
print(car1.__get_model())
````

````python
# Access a private attribute or method using a name mangling syntax
print(car1._Car__model)

print(car1._Car__get_model())
````

1. Write a Python class to represent points and calculate the distance between two points.

image.png

````python
import math
class Point:
  # Code here
  def __init__(self, x: int | float, y: int | float) -> None:
    if (type(x) is not int or type(y) is not int) and (type(y) is not float or type(y) is not float):
      raise TypeError("x and y must be int or float type") 
    self.x = x
    self.y = y

  def distance_to(self, p):
    if not isinstance(p, Point):
      raise TypeError(f"p need to be Point but got {type(p).__name__}")
    return ((p.x - self.x)**2 + (p.y - self.y)**2)**(1/2)

p1 = Point(0,1)
p2 = Point(2,2)
print(p1.distance_to(p2))
````

  2. Write a Python class to represent animals that have different  speak voice. For example Dog-bokbok, Cat-meow, Bird-jibjib. Each class should have at least one attribute and one method.

````python
# Code here
class Animal:
   def __init__(self, name: str, speakWord: str) -> None:
      self.name = name
      self.speakWord = speakWord 
   def speak(self):
      if not self.name: raise NotImplementedError("Did not specify name yet")
      if not self.speakWord: raise NotImplementedError("Did not add speak yet")
      return f"{self.name} '{self.speakWord}'"

class Dog(Animal):
   def __init__(self, name: str) -> None:
      super().__init__(name, speakWord="Bokbok!")

class Cat(Animal):
   def __init__(self, name: str) -> None:
      super().__init__(name, speakWord="Meow!")

class Bird(Animal):
   def __init__(self, name: str) -> None:
      super().__init__(name, "Jibjib!")

pet1 = Dog('น้องหมา')
pet2 = Cat('น้องแมว')
pet3 = Bird('น้องนก')
print(pet1.speak())
print(pet2.speak())
print(pet3.speak())

animals = [pet1, pet2, pet3]

for ani in animals:
   print(ani.speak())
````
