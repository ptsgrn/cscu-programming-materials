```python
def quad1(a, b, c):
    return (-1 * b + (b ** 2 - 4 * a * c) ** (1/2) ) / (2 * a)

def quad2(a, b, c):
    return (-1 * b - (b ** 2 - 4 * a * c) ** (1/2)) / (2 * a)

def run():
    filename = input("Choose your problem file:")
    lines = []
    with open(filename) as f:
        lines = [n.strip().split(" ") for n in f.readlines()]
    for line in lines:
        a, b, c = [int(n) for n in line]
        if a == 0 or (b**2 - 4 * a * c < 0):
            print("Invalid problem")
            continue
        print(quad1(a, b, c), quad2(a, b, c))
```
