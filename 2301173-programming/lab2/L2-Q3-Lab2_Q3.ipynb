## Q3
จงเขียนโปรแกรมที่รับพิกัด (x,y) ของจุดหนึ่ง ในรูปแบบจำนวนเต็ม แล้วตรวจสอบว่าจุดนั้นอยู่ในบริเวณ A หรือ B หรือ C หรือ D ในรูปข้างล่าง 


![xxxx.png](../lab2/files_UploadFile_1_13_Lab2_Q3.png)



ตัวอย่างการทำงาน: 

x,y : **10,20** <br>
( 10 , 20 ) is in A

### ตัวอย่างการทำงาน: 
 
x,y : **0,0** <br>
( 0 , 0 ) is in B

### ตัวอย่างการทำงาน: 

x,y : **9,19** <br>
( 9 , 19 ) is in C

### ตัวอย่างการทำงาน: 

x,y : **20,20** <br>
( 20 , 20 ) is in A

### ตัวอย่างการทำงาน: 

x,y : **40,-10** <br>( 40 , -10 ) is in D

```python
def run():
    A = [(0, 0), (40, 40)]
    B = [(-40, -20), (10, 20)]
    x, y = [int(n) for n in input('x,y :').split(',')]
    isInA = False
    isInB = False
    # If in A area
    if A[0][0] < x < A[1][0] and A[0][1] < y < A[1][1]:
        isInA = True
    if B[0][0] < x < B[1][0] and B[0][1] < y < B[1][1]:
        isInB = True
    print(f"( {x} , {y} ) is in", "D" if not isInA and not isInB else "C" if isInA and isInB else "A" if isInA else "B" if isInB else "????")
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
# run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO
import math

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['10,20'],expected_output='(')
print('ok')
```

```python
'''ตรวจสอบผลการคำนวณและการแสดงผลลัพธ์ให้เหมือนกับโจทย์ที่กำหนดให้'''
test(['10,20'],'( 10 , 20 ) is in A\n')  #A
test(['20,20'],'( 20 , 20 ) is in A\n')  #A
print('ok')
```

```python
test(['0,0'],'( 0 , 0 ) is in B\n')    #B
test(['0,10'],'( 0 , 10 ) is in B\n')  #B
print('ok')
```

```python
test(['9,19'],'( 9 , 19 ) is in C\n')   #C 
test(['1,2'],'( 1 , 2 ) is in C\n')   #C 
print('ok')
```

```python
test(['40,-10'],'( 40 , -10 ) is in D\n') #D
test(['0,20'],'( 0 , 20 ) is in D\n') #D
test(['10,0'],'( 10 , 0 ) is in D\n') #D
print('ok')
```
