## Q6
จงเขียนโปรแกรมที่รับน้ำหนัก และส่วนสูง จากผู้ใช้ มาคำนวณ bmi โดยให้ระบุหน่วยของน้ำหนักเป็น kg
หรือ lbs และ หน่วยของส่วนสูงเป็น cm หรือ m หรือ ft แล้วไปแปลงเป็น kg และ m และคำนวณแสดงผลลัพธ์ว่าผู้ใช้มีรูปร่างในระดับใด <br>
(1 kg. = 2.205 lbs, 1 m. = 3.2808399 feet)

### ตัวอย่างการทำงาน: 

Weight : **61.3 kg**<br>
Height : **1.65 m**<br>
รูปร่างปกติ

### ตัวอย่างการทำงาน: 

Weight : **159.3 lbs** <br>
Height : **4.92 ft**<br>
อ้วนระดับ 2

### ตัวอย่างการทำงาน:

Weight : **110.8 lbs** <br> Height : **167 cm**<br>ผอม


```python
import sys

def run():
    LBS = 1 / 2.205
    FEET = 1 / 3.2808399
    CM = 1 / 100
    w, wu = input("Weight :").split(" ")
    w = float(w)
    w = w * LBS if wu == "lbs" else w
    h, hu = input("Height :").split(" ")
    h = float(h)
    h = h * FEET if hu == "ft" else h * CM if hu == "cm" else h

    bmi = w / h ** 2
    print(w, wu, h, hu, file=sys.stderr)
    print(bmi, file=sys.stderr)
    if bmi < 18.5:
        print("ผอม")
    elif 18.5 <= bmi < 23:
        print("รูปร่างปกติ")
    elif 23 <= bmi < 25:
        print("รูปร่างอ้วน")
    elif 25 <= bmi < 30:
        print("อ้วนระดับ 1")
    else:
        print("อ้วนระดับ 2")
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
# run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['61.3 kg','1.65 m'],expected_output='รูปร่าง')
print('ok')
```
