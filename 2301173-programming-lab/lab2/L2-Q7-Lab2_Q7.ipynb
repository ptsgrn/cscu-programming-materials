เลขหลักที่ 3 (จากหน้า) ของรหัสนิสิตระดับปริญญาบัณฑิต คือ 3, 4 และบัณฑิตศึกษา คือ 7 <br>
เลข 2 หลักสุดท้ายของรหัสนิสิตเป็นรหัสคณะ <br>
รหัสคณะในกลุ่ม 1 คือ 21 23 25 28 <br>
รหัสคณะในกลุ่ม 2 คือ 22 24 26 27 <br>
และแสดงผลลัพธ์ดังตัวอย่างข้างล่างนี้

### ตัวอย่างการทำงาน:

Registration fee : 21000

### ตัวอย่างการทำงาน:

### ตัวอย่างการทำงาน:

Enter semester : **3**  <br>

### ตัวอย่างการทำงาน:

### ตัวอย่างการทำงาน:

### ตัวอย่างการทำงาน:

```python
import sys
def run():
    id = input("Enter student ID :")
    if not (
      int(id[0:2]) >= 50
      and len(id) == 10
      and int(id[2]) in [3,4,7]
      and int(id[:2]) in range(50, 67) 
      and int(id[2]) in [3,4,7]
      and int(id[8:11]) in range(21, 29)
    ):
      print("Invalid ID")
      return

    student_year = "old" if 50 <= int(id[0:2]) <= 55 else "new"
    semester = int(input("Enter semester :"))
    if not semester in range(1,4):
      print("Invalid semester")
      return
    semester = semester in [1,2] and "normal" or "summer"
    status = int(id[2]) in [3,4] and "u" or "g"
    faculty = int(id[8:11]) in [21, 23, 25, 28] and "group1" or "group2"
    print(f"{student_year=} {faculty=} {status=} {semester=}", file=sys.stderr)
    
    feeData = {
      "old": {
        "group1": {
          "u": {
            "normal": 18000,
            "summer": 4500
          },
          "g": {
            "normal": 26000,
            "summer": 7000
          }
        },
        "group2": {
          "u": {
            "normal": 14500,
            "summer": 4500
          },
          "g": {
            "normal": 19000,
            "summer": 7000
          }
        }
      },
      "new": {
        "group1": {
          "u": {
            "normal": 21000,
            "summer": 5250
          },
          "g": {
            "normal": 31000,
            "summer": 7750
          }
        },
        "group2": {
          "u": {
            "normal": 17000,
            "summer": 5250
          },
          "g": {
            "normal": 23000,
            "summer": 7750
          }
        }
      },
    }
    print(f"Registration fee : {feeData.get(student_year, {}).get(faculty, {}).get(status, {}).get(semester, False)}")
    
#    เลขหลักที่ 3 (จากหน้า) ของรหัสนิสิตระดับปริญญาบัณฑิต คือ 3, 4 และบัณฑิตศึกษา คือ 7 <br>
# เลข 2 หลักสุดท้ายของรหัสนิสิตเป็นรหัสคณะ <br>
# รหัสคณะในกลุ่ม 1 คือ 21 23 25 28 <br>
# รหัสคณะในกลุ่ม 2 คือ 22 24 26 27 <br>
# และแสดงผลลัพธ์ดังตัวอย่างข้างล่างนี้
# Enter student ID : **5443200022**  <br>
# Enter semester : **0**  <br>
# Invalid semester
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['5233235425','2'],expected_output='Registration')
print('ok')
```
